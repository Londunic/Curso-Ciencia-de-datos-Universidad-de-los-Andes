### 1. Importación de librerías y archivos

```python
# Importar las librerías necesarias según el análisis que se vaya a realizar
# Librería para manejo de datos
import pandas as pd
# Librería para manejo de condicionales
import numpy as np
```

```python
#Cargar los datos de las órdenes
data=pd.read_excel('empleadosTecnologiaAlpes.xlsx', sheet_name='empleadosTecnologíaAlpes')
```

### 4. Uso de función Where

```python
#Se carga los datos de los departamentos
departamentos= pd.read_excel('empleadosTecnologiaAlpes.xlsx', sheet_name="Departamentos")
#Se le cambia el nombre a la columna ID.
departamentos = departamentos.rename({'ID': 'ID Departamento'}, axis=1)
departamentos.head()
```

```python
#Se realiza la union entre la informacion de empleados y departamentos
union_data=pd.merge(data,departamentos,on="ID Departamento",how='inner')
union_data
```

```python
union_data
```

```python
# asignarlo a una columna nueva de la fuente de datos
union_data['Adm_o_tec']=np.where(union_data['Departamento']=='Investigacion & Desarrollo', 'Tecnico','Administrativo')
# visualizar la columna nueva
union_data[['Adm_o_tec']].head(30)
```
