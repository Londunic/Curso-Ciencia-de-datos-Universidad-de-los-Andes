### 2. Merge 

* Unión interna (Inner Join).
* Unión izquierda (Left Join).
* Unión derecha (Rigth Join).
* Unión completa (Full Outer Join).

```python
#Cambio en el nombre de los ID.
roles = roles.rename({'ID': 'Rol ID'}, axis=1)
roles
```

```python
#Cambio en el nombre de los ID.
departamentos = departamentos.rename({'ID': 'ID Departamento'}, axis=1)
departamentos.head()
```

```python
# Realizar la unión con la función merge
data_total=pd.merge(
     empleados, #Base izquierda
     roles, # Base derecha
     on='Rol ID', #llave a través de la cual vamos a unir
     how='inner' # Forma de realizar el JOIN en este caso: Unión interna
         )
#Visualizar la unión
data_total
```
